```python
import numpy as np
import matplotlib
# Show Plot in The Notebook
matplotlib.use("nbagg")
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
import time
```

```python
CFL = 0.5   #1   # to who dispersion 0.5
dx = 5 #m  #1     # to who dispersion 5
c = 2000 #m/s  #2000
dt = (CFL/c)*dx
nx = 5000 #number of spatial points
Tmax = 8 #max time used 2 for speed sake
t = np.arange(0,Tmax+dt,dt)
nt = len(t)
fdom = 20 #Hz
t0 = 1./fdom
lamb = c/fdom #wavelength

print(f" Velocity:{c} \n \
Dom Freq:{fdom} \n \
Wavelength: {lamb}")
Tlong=np.linspace(0,Tmax,nx)
t.shape
dt
```

Create a ricker wavelet with peak frequency $f_{dom}$

$$ f(t)=(1-2 \pi ^{2}f_{dom}^{2}t^{2}) e^{-\pi ^{2}f_{dom}^{2}t^{2}} $$

```python
#Ricker Source
tmp = np.pi**2 * fdom**2 * (t - t0)**2
ricker = (1 - 2* tmp)* np.exp(-tmp)
```

```python
#Initialize time displacement vectors
u = np.zeros(nx)
unew = np.zeros(nx)
uold = np.zeros(nx)
d2px = np.zeros(nx)
reflect = True
# #initial conditions
# u[0] = 0; u[-1] = 0
M_1 = np.zeros([len(t),nx])
start = time.time()
for n in range(nt):

    for j in range(1,nx-1):
        d2px[j] =  (u[j+1] - 2*u[j] + u[j-1]) /dx**2

    unew = (d2px*dt**2 * c**2) + 2*u - uold
    unew[1] += (dt**2 * ricker[n]) /dx

    # if reflect ==True:
    #     unew[0] = 0; unew[1000] = 0   ##### Reflecting boundary
    M_1[n,:] = unew
    uold, u = u, unew
end = time.time()
total_time_a1 = (end-start) / 60.
print(f"3-point 1D time: {total_time_a1} minutes")
```

```python
fig,ax1 = plt.subplots(1,figsize=(12,6))
def update(frame):
    ax1.clear()
    ax1.grid()
    ax1.plot(M_1[frame,:],'k',linewidth=1.5)
    #ax1.vlines(1000,np.min(M_1),np.max(M_1),color='r',linewidth=2)
    ax1.set_ylim(np.min(M_1),np.max(M_1))
    ax1.set_xlim(0,800)
    ax1.set_title(f'Travelling Ricker Wavelet \n \
       No Boundary \
        time = {np.round(frame*dt,2)}\
        CFL:{CFL}, dx:{dx}, dt:{dt},c:{c}')
    
ani = animation.FuncAnimation(fig,update, frames = np.arange(0,len(t),100),interval=50) #40
ani.save('Ricker_Dispersion_2nd_algorithm1.gif', writer = 'pillow', fps = 4) #3
plt.show()
```

```python
#####4th order
#Initialize time displacement vectors
u = np.zeros(nx)
unew = np.zeros(nx)
uold = np.zeros(nx)
d2px = np.zeros(nx)

M_1_4 = np.zeros([len(t),nx])
start = time.time()

for i in range(nt):   #loop through t

    for j in range(2,nx-2):
        d2px[j] = ((-1./12)*u[j+2] +(4./3)* u[j+1] - (5./2)*u[j] +(4./3)*u[j-1] +(-1./12)*u[j-2]) / dx**2

    unew = c**2 * dt**2 *d2px + 2*u - uold #Calculate next time step
##############
######Error here
    unew[2] += (dt**2 *(ricker[i]))/dx #Introduce the source
##################
    M_1_4[-i,:] = unew        #Put into matrix for animation
    uold, u = u, unew    #Update all displacements

end = time.time()
total_time_a4 = (end-start) / 60.
print(f"5-point 1D time: {total_time_a4}")

```

```python
def fourier_derivative(f,dx):
    Nx = int(len(f))
    kmax = np.pi/dx
    dk = kmax/(Nx/2)
    k = np.arange(float(Nx))
    k[:Nx//2] *= dk
    k[Nx//2:] -= kmax

    ff = 1j*k *np.fft.fft(f)
    df = np.real(np.fft.ifft(ff))
    #df = (df-min(df))/ (max(df)- min(df))
    return df

def fourier_derivative_2nd(f,dx):
    Nx = int(len(f))
    kmax = np.pi/dx
    dk = kmax/(Nx/2)
    k = np.arange(float(Nx))
    k[:Nx//2] *= dk
    k[Nx//2:] -= kmax

    ff = (1j*k)**2 *np.fft.fft(f)
    df = np.real(np.fft.ifft(ff))
    #df = (df-min(df))/ (max(df)- min(df))
    return df

ricker_der1 = fourier_derivative(ricker,dx)
ricker_der2 = fourier_derivative_2nd(ricker,dx)

```
